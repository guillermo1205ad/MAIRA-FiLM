```python
import os
import json
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from PIL import Image
from tqdm import tqdm

torch.backends.cudnn.enabled = False  # Evitar problemas con GRU

# ------------------- Verificación de rutas -------------------

def verify_directories(base_dir):
    required_paths = [
        os.path.join(base_dir, 'images/trainA'),
        os.path.join(base_dir, 'images/valA'),
        os.path.join(base_dir, 'images/valB'),
        os.path.join(base_dir, 'questions/CLEVR_trainA_questions.json'),
        os.path.join(base_dir, 'questions/CLEVR_valA_questions.json'),
        os.path.join(base_dir, 'questions/CLEVR_valB_questions.json')
    ]
    for path in required_paths:
        if not os.path.exists(path):
            raise FileNotFoundError(f"No se encontró el archivo o directorio requerido: {path}")
    print("Todas las rutas requeridas existen.")

# ------------------- Checkpoint Functions -------------------

def save_checkpoint(state, filename='checkpoint.pth'):
    torch.save(state, filename)
    print(f"Checkpoint guardado en {filename}")

def save_best_model(model, filename='best_film_model_finetuning.pth'):
    torch.save(model.state_dict(), filename)
    print(f"Mejor modelo guardado en {filename}")

def load_checkpoint(filename='best_film_model.pth', model=None, optimizer=None):
    if os.path.isfile(filename):
        print(f"Cargando checkpoint desde {filename}...")
        checkpoint = torch.load(filename)
        if 'model_state_dict' in checkpoint:
            model.load_state_dict(checkpoint['model_state_dict'])
            if optimizer and 'optimizer_state_dict' in checkpoint:
                optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
            start_epoch = checkpoint.get('epoch', 0) + 1
            batch_idx = checkpoint.get('batch_idx', 0)
            best_accuracy = checkpoint.get('best_accuracy', 0.0)
            print(f"Reanudando desde la época {start_epoch}, batch {batch_idx}, con precisión {best_accuracy:.2f}%")
        else:
            model.load_state_dict(checkpoint)
            print("Modelo cargado con solo los pesos. Comenzando desde la época 0.")
            start_epoch, batch_idx, best_accuracy = 0, 0, 0.0
        return model, optimizer, start_epoch, batch_idx, best_accuracy
    else:
        print("No se encontró un checkpoint. Iniciando desde la época 0.")
        return model, optimizer, 0, 0, 0.0

# ------------------- Modelo y Dataset -------------------

class FiLMResBlock(nn.Module):
    def __init__(self, in_channels):
        super(FiLMResBlock, self).__init__()
        self.conv1 = nn.Conv2d(in_channels, in_channels, kernel_size=1)
        self.conv2 = nn.Conv2d(in_channels, in_channels, kernel_size=3, padding=1)
        self.relu = nn.ReLU()

    def forward(self, x, gamma, beta):
        out = self.conv1(x)
        out = self.conv2(self.relu(out))
        gamma = gamma.unsqueeze(2).unsqueeze(3)
        beta = beta.unsqueeze(2).unsqueeze(3)
        out = gamma * out + beta
        return self.relu(out)

class FiLMNetwork(nn.Module):
    def __init__(self, num_classes, max_length, use_resnet=True):
        super(FiLMNetwork, self).__init__()
        if use_resnet:
            resnet = models.resnet101(weights=models.ResNet101_Weights.DEFAULT)
            for param in resnet.parameters():
                param.requires_grad = False
            self.feature_extractor = nn.Sequential(
                resnet.conv1,
                resnet.bn1,
                resnet.relu,
                resnet.maxpool,
                resnet.layer1,
                resnet.layer2,
                resnet.layer3
            )
            feature_channels = 1024
        else:
            self.feature_extractor = nn.Sequential(
                nn.Conv2d(3, 128, kernel_size=4, stride=2, padding=1),
                nn.BatchNorm2d(128),
                nn.ReLU(),
                nn.Conv2d(128, 128, kernel_size=4, stride=2, padding=1),
                nn.BatchNorm2d(128),
                nn.ReLU()
            )
            feature_channels = 128

        self.reduce_channels = nn.Conv2d(feature_channels + 2, 128, kernel_size=1)
        self.resblocks = nn.ModuleList([FiLMResBlock(128) for _ in range(4)])
        self.word_embedding = nn.Embedding(256, 200)
        self.gru = nn.GRU(input_size=200, hidden_size=4096, batch_first=True)
        self.affine_gamma = nn.Linear(4096, 128)
        self.affine_beta = nn.Linear(4096, 128)
        self.classifier = nn.Sequential(
            nn.Conv2d(128, 512, kernel_size=1),
            nn.ReLU(),
            nn.AdaptiveMaxPool2d(1),
            nn.Flatten(),
            nn.Linear(512, 1024),
            nn.ReLU(),
            nn.Linear(1024, num_classes)
        )

    def add_coordinates(self, features):
        batch_size, _, height, width = features.size()
        x_coords = torch.linspace(-1, 1, steps=width).repeat(batch_size, height, 1).unsqueeze(1)
        y_coords = torch.linspace(-1, 1, steps=height).repeat(batch_size, width, 1).unsqueeze(1).transpose(2, 3)
        coords = torch.cat([x_coords, y_coords], dim=1).to(features.device)
        return torch.cat([features, coords], dim=1)

    def forward(self, images, questions):
        features = self.feature_extractor(images)
        features = self.add_coordinates(features)
        features = self.reduce_channels(features)
        question_embedding = self.gru_forward(questions)
        gamma = self.affine_gamma(question_embedding)
        beta = self.affine_beta(question_embedding)
        for resblock in self.resblocks:
            features = resblock(features, gamma, beta)
        return self.classifier(features)

    def gru_forward(self, questions):
        embedded = self.word_embedding(questions)
        _, hidden = self.gru(embedded)
        return hidden.squeeze(0)

class CLEVRDataset(Dataset):
    def __init__(self, image_dir, question_file, max_length, transform=None):
        if not os.path.exists(image_dir):
            raise FileNotFoundError(f"La carpeta de imágenes no existe: {image_dir}")
        if not os.path.exists(question_file):
            raise FileNotFoundError(f"El archivo de preguntas no existe: {question_file}")

        self.image_dir = image_dir
        with open(question_file, 'r') as f:
            self.questions = json.load(f)['questions']
        self.transform = transform
        self.max_length = max_length

    def __len__(self):
        return len(self.questions)

    def __getitem__(self, idx):
        question = self.questions[idx]
        image_path = os.path.join(self.image_dir, question['image_filename'])
        if not os.path.exists(image_path):
            raise FileNotFoundError(f"La imagen no existe: {image_path}")
        image = Image.open(image_path).convert('RGB')
        question_text = question['question'][:self.max_length]
        question_tensor = torch.tensor([ord(c) for c in question_text], dtype=torch.long)
        if question_tensor.size(0) < self.max_length:
            question_tensor = torch.nn.functional.pad(question_tensor, (0, self.max_length - question_tensor.size(0)))
        label = int(question['answer']) if question['answer'].isdigit() else 0
        if self.transform:
            image = self.transform(image)
        return image, question_tensor, label

def collate_fn(batch):
    images, questions, labels = zip(*batch)
    images = torch.stack(images)
    questions = torch.stack(questions)
    labels = torch.tensor(labels)
    return images, questions, labels

# ------------------- Función para Cargar Múltiples Conjuntos de Datos -------------------

def load_data():
    base_dir = '/home/gperaltag/mi_entorno/data/mnt/ialabnas/datasets/CLEVR_CoGenT_v1.0/CLEVR_sample_10_000/sample_2'
    verify_directories(base_dir)
    transform = transforms.Compose([transforms.Resize((224, 224)), transforms.ToTensor()])
    max_length = 100

    trainA = CLEVRDataset(
        os.path.join(base_dir, 'images/trainA'),
        os.path.join(base_dir, 'questions/CLEVR_trainA_questions.json'),
        max_length,
        transform
    )
    valA = CLEVRDataset(
        os.path.join(base_dir, 'images/valA'),
        os.path.join(base_dir, 'questions/CLEVR_valA_questions.json'),
        max_length,
        transform
    )
    valB = CLEVRDataset(
        os.path.join(base_dir, 'images/valB'),
        os.path.join(base_dir, 'questions/CLEVR_valB_questions.json'),
        max_length,
        transform
    )

    train_loader = DataLoader(trainA, batch_size=64, collate_fn=collate_fn, shuffle=True, num_workers=32)
    val_loaderA = DataLoader(valA, batch_size=64, collate_fn=collate_fn, num_workers=32)
    val_loaderB = DataLoader(valB, batch_size=64, collate_fn=collate_fn, num_workers=32)

    return train_loader, val_loaderA, val_loaderB

# ------------------- Función de Entrenamiento con Early Stopping -------------------

def train(model, train_loader, val_loaderA, val_loaderB, criterion, optimizer, device, epochs, start_epoch=0, best_accuracy=0.0, patience=3):
    train_losses, val_lossesA, val_lossesB = [], [], []
    train_accuracies, val_accuraciesA, val_accuraciesB = [], [], []
    early_stop_counter = 0

    for epoch in range(start_epoch, epochs):
        model.train()
        epoch_loss, correct_train, total_train = 0, 0, 0

        for images, questions, labels in tqdm(train_loader, desc=f"Época {epoch+1}/{epochs} - Entrenando"):
            images, questions, labels = images.to(device), questions.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images, questions)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            epoch_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            correct_train += (predicted == labels).sum().item()
            total_train += labels.size(0)

        train_losses.append(epoch_loss / len(train_loader))
        train_accuracies.append(100 * correct_train / total_train)

        # Validación en valA
        model.eval()
        val_lossA, correct_valA, total_valA = 0, 0, 0
        with torch.no_grad():
            for images, questions, labels in tqdm(val_loaderA, desc="Validando en valA"):
                images, questions, labels = images.to(device), questions.to(device), labels.to(device)
                outputs = model(images, questions)
                loss = criterion(outputs, labels)
                val_lossA += loss.item()
                _, predicted = torch.max(outputs, 1)
                correct_valA += (predicted == labels).sum().item()
                total_valA += labels.size(0)
        val_lossesA.append(val_lossA / len(val_loaderA))
        val_accuraciesA.append(100 * correct_valA / total_valA)

        # Validación en valB
        val_lossB, correct_valB, total_valB = 0, 0, 0
        with torch.no_grad():
            for images, questions, labels in tqdm(val_loaderB, desc="Validando en valB"):
                images, questions, labels = images.to(device), questions.to(device), labels.to(device)
                outputs = model(images, questions)
                loss = criterion(outputs, labels)
                val_lossB += loss.item()
                _, predicted = torch.max(outputs, 1)
                correct_valB += (predicted == labels).sum().item()
                total_valB += labels.size(0)
        val_lossesB.append(val_lossB / len(val_loaderB))
        val_accuraciesB.append(100 * correct_valB / total_valB)

        print(f"Epoch [{epoch+1}/{epochs}] - Accuracy ValA: {val_accuraciesA[-1]:.2f}%, Accuracy ValB: {val_accuraciesB[-1]:.2f}%")

        # Early stopping
        if val_accuraciesA[-1] > best_accuracy:
            best_accuracy = val_accuraciesA[-1]
            save_best_model(model)
            early_stop_counter = 0
        else:
            early_stop_counter += 1

        if early_stop_counter >= patience:
            print("Early stopping triggered")
            break

    # Gráficos
    epochs_range = range(start_epoch + 1, epoch + 2)
    plt.figure(figsize=(12, 5))
    plt.plot(epochs_range, train_losses, label="Pérdida de Entrenamiento")
    plt.plot(epochs_range, val_lossesA, label="Pérdida de Validación valA")
    plt.plot(epochs_range, val_lossesB, label="Pérdida de Validación valB")
    plt.xlabel("Épocas")
    plt.ylabel("Pérdida")
    plt.title("Pérdida durante el Entrenamiento y la Validación")
    plt.legend()
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(12, 5))
    plt.plot(epochs_range, train_accuracies, label="Precisión de Entrenamiento")
    plt.plot(epochs_range, val_accuraciesA, label="Precisión de Validación valA")
    plt.plot(epochs_range, val_accuraciesB, label="Precisión de Validación valB")
    plt.xlabel("Épocas")
    plt.ylabel("Precisión (%)")
    plt.title("Precisión durante el Entrenamiento y la Validación")
    plt.legend()
    plt.grid(True)
    plt.show()

# ------------------- Función Principal de Fine-tuning -------------------

def fine_tune():
    train_loader, val_loaderA, val_loaderB = load_data()
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = FiLMNetwork(num_classes=28, max_length=100).to(device)
    criterion = nn.CrossEntropyLoss()

    for name, param in model.named_parameters():
        if 'feature_extractor' in name:
            param.requires_grad = False
        else:
            param.requires_grad = True

    optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=3e-4, weight_decay=1e-5)
    model, optimizer, start_epoch, batch_idx, best_accuracy = load_checkpoint('best_film_model.pth', model, optimizer)
    train(model, train_loader, val_loaderA, val_loaderB, criterion, optimizer, device, epochs=3, start_epoch=start_epoch, best_accuracy=best_accuracy)

if __name__ == "__main__":
    fine_tune()
```
